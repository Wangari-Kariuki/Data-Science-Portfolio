**B** The name, ID, and closest approach distance (in km) of the asteroid that will come the closest to Earth during your next birthday week.

```python
distance_in_km = []
for index_dict in adf['close_approach_data']:
  # Check if the list is not empty before trying to access its elements
  if len(index_dict) > 0:
    #miss_dist is a dictionary
    dist = index_dict[0]['miss_distance']['kilometers']
    distance_in_km.append(dist)
  else:
    distance_in_km.append(None) # Append None or handle cases where close_approach_data is empty
```

```python
new_adf['distance_in_km'] = distance_in_km
closest = new_adf.loc[new_adf['distance_in_km'].idxmin()]
farthest = new_adf.loc[new_adf['distance_in_km'].idxmax()]
print("closest asteroid to earth: \n ", closest )
print("farthest asteroid to earth: \n ", farthest )
```
